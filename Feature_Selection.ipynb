Condensing Data

```python
#Correlate each colum with each other and remove such features
# First I remove data that has less that 15% of outputs
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
threshold = 0.05
selector = VarianceThreshold(threshold=threshold)
selector.fit(df_cleaned_features)

cols_to_keep = selector.get_support(indices=True)
df_features_condensed = df_cleaned_features.iloc[:, cols_to_keep].copy()
original_count = df_cleaned_features.shape[1]
new_count = df_features_condensed.shape[1]
print(f"Original Feature Count: {original_count}")
print(f"New Condensed Feature Count (after removing sparse/constant columns): {new_count}")
```

```python
df_features_condensed.columns
```

```python
#Hot encoded true and false
import pandas as pd
import numpy as np

# Assuming df_features_condensed is loaded and safe (e.g., used .copy())

# --- 1. CLEAN STRINGS ACROSS ALL OBJECT COLUMNS (Stage 1) ---
object_cols = df_features_condensed.select_dtypes(include=['object']).columns

print(f"Cleaning {len(object_cols)} object columns...")
for col in object_cols:
    # This step is correct: convert to string, strip whitespace, and convert to UPPERCASE
    cleaned_series = df_features_condensed[col].astype(str).str.strip().str.upper()
    df_features_condensed[col] = cleaned_series

# --- 2. PERFORM GLOBAL REPLACEMENT (Stage 2) ---
# This is performed only once, after all string columns have been uniformly cleaned.
# We only need 'TRUE' and 'FALSE' now since we forced the case above.
print("Performing global boolean replacement...")
df_features_condensed.replace({'TRUE': 1, 'FALSE': 0}, inplace=True)

# --- 3. FINAL COERCE TO NUMERIC AND IMPUTATION (Stage 3) ---
# Now, iterate through ALL columns to ensure they are numeric.
print("Final conversion to numeric...")
for col in df_features_condensed.columns:
    df_features_condensed[col] = pd.to_numeric(
        df_features_condensed[col],
        errors='coerce'
    ).astype(np.float32)

df_features_condensed.fillna(0, inplace=True)

print("\nSUCCESS! Data is now fully numerical and ready for the next step.")
```

```python
df_features_condensed.head()
```

```python
# Drop unslaced features
unscaled_features_to_drop= [
    'interest_in_entrepreneurship',
    'interest_social_entr',
    'interest_cooperative_business'
]
redundant_ohe_to_drop = [
    "family_attended_college_No, I don't think so",
    'device_access_Smartphone',
]

df_features_condensed.drop(columns= unscaled_features_to_drop, inplace=True, errors= 'ignore')
df_features_condensed.drop(columns=redundant_ohe_to_drop, inplace=True, errors='ignore')
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.multioutput import MultiOutputClassifier
from sklearn.feature_selection import SelectFromModel

X_159_features= df_features_condensed.copy()
Y= df_targets.copy()

base_estimator= LogisticRegression(
    penalty='l1',
    C=0.1,
    solver= 'liblinear',
    random_state=42
)
selector_model= MultiOutputClassifier(base_estimator, n_jobs=-1)

selector_model.fit(X_159_features, Y)

all_coefs= [est.coef_[0] for est in selector_model.estimators_]
all_coefs_array= np.array(all_coefs)

avg_abs_coef= np.mean(np.abs(all_coefs_array), axis=0)
NUM_TOP_FEATURES= 15
top_feature_indices= np.argsort(avg_abs_coef)[::-1][:NUM_TOP_FEATURES]
selected_features= X_159_features.columns[top_feature_indices].tolist()

X_final_features= X_159_features[selected_features]
final_count= X_final_features.shape[1]

print("\n---- Final Feature Selection Results----")
print(f"Features Selected by Lasso {final_count}")
print(f"Selected Feature Names:\n{selected_features}")


X_final_features.to_csv('X_final_optimized.csv', index=False)
files.download('X_final_optimized.csv')
```

```python
#Add the scaled features and removing redundant features
scaled_features_to_add=[
    'interest_in_entrepreneurship_scaled',
    'interest_social_entr_scaled',
    'interest_cooperative_business_scaled'
]

redundant2_ohe_to_drop= [
    'racial_ethnic_Prefer not to say',
    'racial_ethnic_African American / Black, Hispanic/Latinx',
    'social_welfare_public_housing_None of these apply',
]


for feature in scaled_features_to_add:
  if feature not in selected_features:
    selected_features.append(feature)


for feature in redundant2_ohe_to_drop:
    if feature in selected_features:
        selected_features.remove(feature)

X_final_features = X_159_features[selected_features]
final_count = X_final_features.shape[1]


print("\n---- Final ENFORCED Feature Selection Results----")
print(f"Features Selected by Lasso (Corrected): {final_count}")
print(f"Selected Feature Names:\n{selected_features}")


X_final_features.to_csv('X_final_optimized.csv', index=False)
files.download('X_final_optimized.csv')
```
