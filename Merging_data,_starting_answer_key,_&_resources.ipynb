```python
#one hot encode/tagging
ohe_map = {
    'is_NYCHA_Resident': ['NYCHA resident'],
    'is_NOT_Scaling_Stage': ['NOT Scaling Stage'],
    'is_Idea_Stage': ['Idea Stage', 'At least Idea Stage'],
    'is_NOT_Idea_Stage': ['NOT Idea stage', 'NOT Idea Stage', 'NOT idea sstage'],
    'is_Legally_Registered': ['Legally registered business', 'legally registered business'],
    'is_NO_Revenue': ['NO revenue'],
    'is_YES_Revenue': ['YES revenue'],
    'is_Revenue_Less_Than_5K': ['less than $5,000 in revenue'],
    'is_Revenue_More_Than_5K': ['more than $5,000 in revenue'],
    'is_Revenue_More_Than_500_Last_3Mo': ['more than $500 in revenue over the past 3 months'],
    'is_Vet_or_Family_of_Vet': ['Vet or family of a Vet', 'veteran'],
    'is_Age_18_Plus': ['18 yrs old'],
    'is_Business_in_NYC': ['business inside NYC'],
    'is_On_Lease_Rent_Payments': ['on lease & rent payments'],
    'is_Female': ['Female'],
    'is_Age_50_Plus': ['Older than 50'],
    'is_Tech_Idea': ['Tech-based Idea'],
    'is_Operating_At_Least_1_Year': ['operating for at least a year'],
    'is_Identifies_As_Black': ['Identify as Black'],
    'is_Cannabis_Idea': ['cannabis-based idea'],
    'is_Brooklyn_Resident': ['Brooklyn resident'],
    'is_BKPL_Cardholder': ['BKPL Cardholder']
}

def apply_ohe(df, ohe_map):
    for col in ohe_map.keys():
        df[col] = 0

    for index, row in df.iterrows():
        eligibility = row['eligibility_requirements']

        for ohe_col, search_strings in ohe_map.items():
            is_present = False
            for search_str in search_strings:
                # Use case-insensitive search
                if search_str.lower() in eligibility.lower():
                    is_present = True
                    break

            if is_present:
                df.loc[index, ohe_col] = 1

    return df

df_program_requirements = apply_ohe(df_program_requirements, ohe_map)
df_program_requirements = df_program_requirements.drop(columns=['eligibility_requirements'])
```
