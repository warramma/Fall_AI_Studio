```python
import pandas as pd
from io import StringIO
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import jaccard_score
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from google.colab import files #for google colab
import io
import os
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import features
uploaded = files.upload()
for filename in uploaded.keys():
      df_features = pd.read_csv(io.BytesIO(uploaded[filename]))
      print(f"DataFrame from {filename}:")
```

```python
#import y variable
uploaded = files.upload()
for filename in uploaded.keys():
      df_key = pd.read_excel(io.BytesIO(uploaded[filename]))
      print(f"DataFrame from {filename}:")
```

```python
df_features.head()
```

```python
df_key.head()
```

```python
# feature scaling
X = df_features
y = df_key
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)
```

```python
X_scaled_df.head()
```

```python
#split data for training
X_train, X_test, Y_train, Y_test = train_test_split(
    X_scaled_df, y, test_size=0.3, random_state=42
)

print(f"Training samples: {X_train.shape[0]}")
print(f"Test samples: {X_test.shape[0]}")
```

# KNN

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, Y_train)

print("\nK-Nearest Neighbors Model (k=3) Training Complete.")
```

```python
Y_pred = knn.predict(X_test)
```

```python
jaccard = jaccard_score(Y_test, Y_pred, average='samples')

print(f"\nEvaluation on Test Set (Jaccard Score): {jaccard:.4f}")
```

```python
target_names = y.columns.tolist()

comparison_data = []

for i in range(min(10, len(Y_test))):
    actual_resources_mask = Y_test.iloc[i].values == 1
    actual_resources = [target_names[j].replace('target_', '') for j in np.where(actual_resources_mask)[0]]

    predicted_resources_mask = Y_pred[i] == 1
    predicted_resources = [target_names[j].replace('target_', '') for j in np.where(predicted_resources_mask)[0]]

    comparison_data.append({
        'Actual Resources': ", ".join(actual_resources) if actual_resources else "None",
        'Predicted Resources': ", ".join(predicted_resources) if predicted_resources else "None"
    })

comparison_df = pd.DataFrame(comparison_data)

print("\nSample of True vs. Predicted Resources (First 10 Test Samples):")
print(comparison_df.to_string(index=False))
```

## grid search

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

print("Starting Grid Search for optimal k...")

param_grid = {'n_neighbors': np.arange(1, 15)}
grid_search = GridSearchCV(
    estimator=KNeighborsClassifier(),
    param_grid=param_grid,
    scoring='jaccard_weighted', #for multi label binarizer
    cv=5,
    verbose=1,
    n_jobs=-1
)

grid_search.fit(X_train, Y_train)

best_k = grid_search.best_params_['n_neighbors']
best_score = grid_search.best_score_

print("\n--- Grid Search Results ---")
print(f"Optimal k value found: **{best_k}**")
print(f"Best cross-validation Jaccard Score: **{best_score:.4f}**")
```

```python
optimal_k = best_k

print(f"\nTraining Final KNN Model with optimal k = {optimal_k}...")
final_knn = KNeighborsClassifier(n_neighbors=optimal_k)

final_knn.fit(X_train, Y_train)

Y_pred_final = final_knn.predict(X_test)

jaccard_final = jaccard_score(Y_test, Y_pred_final, average='samples')

print(f"\nFinal Evaluation on Test Set (Optimal k={optimal_k}):")
print(f"Jaccard Score: **{jaccard_final:.4f}**")


target_names = Y_test.columns.tolist()
comparison_data_final = []

for i in range(min(10, len(Y_test))):
    actual_resources_mask = Y_test.iloc[i].values == 1
    actual_resources = [target_names[j].replace('target_', '') for j in np.where(actual_resources_mask)[0]]

    predicted_resources_mask = Y_pred_final[i] == 1
    predicted_resources = [target_names[j].replace('target_', '') for j in np.where(predicted_resources_mask)[0]]

    comparison_data_final.append({
        'Actual Resources': ", ".join(actual_resources) if actual_resources else "None",
        'Predicted Resources': ", ".join(predicted_resources) if predicted_resources else "None"
    })

comparison_df_final = pd.DataFrame(comparison_data_final)

print("\nSample of True vs. Predicted Resources (Optimal Model):")
print(comparison_df_final.to_string(index=False))
```
